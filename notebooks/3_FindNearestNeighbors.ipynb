# Find the K Nearest Neighbors for a particular image

Here we find the k-nearest neighbors based on image features. We load the pickles dictionary and find the nearest neighbors from that portion of the dataset.

```python
import sys
sys.path.append("../main/")
```

```python
from image_utils import find_k_closest, get_k_nearest_neighbor_image
%load_ext autoreload
%autoreload 2

import random
import matplotlib.pyplot as plt
%matplotlib inline
```

## load the pickle file with feature vectors

```python
import pickle
with open('../main/data/features_dict.pickle', 'rb') as handle:
    features_dict = pickle.load(handle)
```

## examples of finding the k nearest neighbors for a few randomly selected images

```python
# we take an objectid to run nearest-neighbors on
# this means that the leftmost image will be the object itself
# the remaining 4 (left to right) are most similar

features_dict_as_list = list(features_dict.items())

for i in range(5):
    objectid, this_vec = random.choice(features_dict_as_list)
    print("Example #{}: object id = {}".format(i, objectid))
    
    k_closest_list = find_k_closest(this_vec, features_dict)
    image = get_k_nearest_neighbor_image(k_closest_list)
    plt.imshow(image)
    plt.show()
    
```
